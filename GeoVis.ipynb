```python
import math
import pandas as pd
from ast import literal_eval
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import CARTODBPOSITRON
from bokeh.io import show, output_notebook
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import CARTODBPOSITRON
p = figure(x_range=(-13150000, -13155000), y_range=(4040000, 4055000),
           x_axis_type="mercator", y_axis_type="mercator")
p.add_tile(CARTODBPOSITRON)
output_notebook()
show(p)
```

```python
# Lat/long -> mercador
def merc(Coords):
    Coordinates = literal_eval(Coords)
    lat = Coordinates[0]
    lon = Coordinates[1]
    
    r_major = 6378137.000
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)

# Pasadena coords
merc('(34.153666052, -118.12583283)')
```

```python
# Loading data into Dataframe
df = pd.read_html('tcds_list.xls')[1]
df.columns = range(df.shape[1])
df.columns = df.iloc[0]
df = df[1:]

selectHeaders = ['Latitude', 'Longitude', 'Latest']
dfLoc = df[selectHeaders]
```

```python
# Cleaning the data so that the nan values are gone
L_Map = pd.DataFrame(columns=selectHeaders)

for i, row in dfLoc.iterrows():
    if str(row["Longitude"]) != 'nan' and str(row["Latitude"]) != 'nan' and str(row["Latest"]) != 'nan':
        L_Map = L_Map.append(row)

# Adding location column with (Latitude, Longitude)
L_Map['Location'] = L_Map.apply(lambda row: '(' + str(row['Latitude']) + ', ' + str(row['Longitude']) + ')', axis=1)
L_Map
```

```python
# Add mercator x and y coords
L_Map['coords_x'] = L_Map['Location'].apply(lambda x: merc(x)[0])
L_Map['coords_y'] = L_Map['Location'].apply(lambda x: merc(x)[1])
```

```python
p = figure(x_axis_type="mercator", y_axis_type="mercator")
p.add_tile(CARTODBPOSITRON)
p.circle(x = L_Map['coords_x'],
         y = L_Map['coords_y'])
output_notebook()
show(p)
```
